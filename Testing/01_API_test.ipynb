```python
import requests
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
url = 'http://127.0.0.1:8000/predict'
url2 = 'http://127.0.0.1:8000/predict'
url3 = 'https://credit-card-fraud.wonderfulstone-60fe1e35.westus2.azurecontainerapps.io/predict'
```

```python
results = []
for i in range(1000):
    sample =  pd.read_csv('sample.csv').to_dict(orient='records')[i]
    response = requests.post(url3, json=sample)
    #print(f"Response for sample {i}: {response.json()}")
    results.append(response.json()['prediction'])
```

```python
sample_target = pd.read_csv('sample.csv')['target'].tolist()
```

```python
accuracy = accuracy_score(sample_target, results)
print(f"Accuracy of the API predictions: {accuracy:.2f}")
```

```python
sample_df = pd.DataFrame({'target': sample_target, 'prediction': results})
```
