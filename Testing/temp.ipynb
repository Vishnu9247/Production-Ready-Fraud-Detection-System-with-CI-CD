```python
import pandas as pd
```

```python
df = pd.read_csv("../Data Files/Raw Files/credit_card_frauds.csv")
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['year'] = df['trans_date_trans_time'].dt.year
```

```python
df['year'].value_counts()
```

```python
df_2019 = df[df['year'] == 2019]
df_2020 = df[df['year'] == 2020]

df_2019.drop(columns=['year'], inplace=True)
df_2020.drop(columns=['year'], inplace=True)

df_2019.to_csv("../Data Files/Staging/credit_card_frauds_2019.csv", index=False)
df_2020.to_csv("../Data Files/Staging/credit_card_frauds_2020.csv", index=False)
```
