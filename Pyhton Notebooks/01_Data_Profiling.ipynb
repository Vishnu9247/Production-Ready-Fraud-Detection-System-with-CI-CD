### Profiling Data

##### Importing all the required libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

##### Loading data from csv file

```python
df = pd.read_csv('../Data Files/Raw Files Prototype/credit_card_frauds.csv')
df.head()
```

##### Removing spaces and gaps in column names and replacing them with underscore

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(r"[()\-\/]", "", regex=True)
)

print(df.info())
```

##### Checking for any Null or Duplicated values and rows

```python
print(f'Null values: {df.isnull().sum().sum()}')
print(f'Duplicated rows: {df.duplicated().sum()}')
```

##### Finding the number of Distinct and Unique values in a column.

```python
summary = []

for col in df.columns:
    distinct_count = df[col].nunique(dropna=False)
    value_counts = df[col].value_counts(dropna=False)
    unique_count = (value_counts == 1).sum()

    summary.append({
        "Column": col,
        "Distinct Count": distinct_count,
        "Unique Count": unique_count
    })

summary_df = pd.DataFrame(summary)

print(summary_df)
```
